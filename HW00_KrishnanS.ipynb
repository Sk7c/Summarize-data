## 5. Convert Objects to Dates using DateTime

- Convert variables from object datatype to datetime datatype

```python
df_weather['MaxTime'] = pd.to_datetime(df_weather['MaxTime'])
df_weather['MinTime'] = pd.to_datetime(df_weather['MinTime'])
df_weather['sunriseTime'] = pd.to_datetime(df_weather['sunriseTime'])
df_weather['sunsetTime'] = pd.to_datetime(df_weather['sunsetTime'])
df_weather.info()
```

## Summary - Convert Objects to Dates using DateTime

- used to_datetime feature from pandas to convert object datatype to datetime datatype 
- used info() to verify if the datatype changes have been implemented.

## 6. Create additional date features

1. Create a [‘MaxDay’] feature with number for dayofweek
2. Create a [‘MaxMonth’] feature with number for month
3. Create a [‘MaxDayName’] feature with number for day_name( )
4. Show top 5 rows of df_weather

```python
df_weather['MaxDay'] = df_weather['MaxTime'].dt.dayofweek
df_weather['MaxMonth'] = df_weather['MaxTime'].dt.month
df_weather['MaxDayName'] = df_weather['MaxTime'].dt.day_name()
```

```python
#final output with integer datatypes for df_weather['MaxDay'] and df_weather['MaxMonth']
df_weather.head()
```
